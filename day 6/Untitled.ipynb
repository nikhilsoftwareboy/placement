```python
#create a function that return opening web server on port number and its default value is 80 or user can pass custom port number
port=80
def change():
    global port
    port=int(input("Enter custom port : "))
    return port
change()
print(port)
```

```python
#lambda function -> no need to write return , can do in single line with using lambda

add=lambda a,b:a+b
print(add(2,3))
sub=lambda a,b:a-b
print(sub(2,3))
mul=lambda a,b:a*b
print(mul(2,3))
div=lambda a,b:a/b
print(div(2,3))
```

```python
#lambda with conditions
max_val = lambda a,b: a if a>b else b
max_val(1,2)
```

```python
#lambda conditions
num = lambda a: "even" if a%2==0 else "odd"
print(num(2))
```
