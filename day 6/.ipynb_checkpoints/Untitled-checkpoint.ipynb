```python
def sub(a,b):
    return(a-b)
def mul(a,b):
    return(a*b)
def div(a,b):
    return(a/b)
sub(3,2)
mul(3,2)
div(3,2)
```

```python
#global variable
x=5

def update():
    global x
    x=x-2
    print(x)
update()
```
